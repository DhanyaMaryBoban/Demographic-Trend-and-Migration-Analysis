# Final Report : Demographic Trend and Migration Analysis

### 1. Summary
This analysis explores population trends and net migration patterns across five South Asian countries including India, Pakistan, Bangladesh, Sri Lanka and Afghanistan. The findings highlight consistent population growth, negative net migration across all countries and weak correlation between population size and migration. The region emerges as asignificant source of emigrants globally, driven by economic, social and political factors.

### 2. Key Insights
1. **Population Trends**:
- India is the most populous country, showing steady growth from 2019-2023.
- Despite fluctuations, overall population continues to rise, indicating strong natural increase(birth rate > death rate).

2. **Migration Patterns**:
- In the analysis countries shows negative net migration. This means in all countries more people are leaving than entering.
- Bangladesh, Pakistan and India contribute heavily to global migration flows.
- Brief rise in immigration in Afghanistan during the 1980s, linked to geopolitical events.

3. **Correlation Analysis**:
- Weak negative correlation (-0.123) between population size and net migration.
- Migration is influenced more by external factors (economy, politics, conflict) than population size.

4. **Power BI Insights**:
- Net Migration from 2019 to 2023 is -2 million. It indicates that across 2019–2023, the regions experienced a net loss of 2 million people due to migration.
- The population across the given South Asian countries declined by 2.49% per year during the period analyzed.This decline could be due to outward migration, lower birth rates, or data inconsistencies.
- Population trend shows anomalies (2021 data spike requires validation).
- Map visualization confirms India as the largest population, followed by Bangladesh and Pakistan.
   

### 3. Conclusion
South Asian remains a major source of emigrants gobally, with population growth driven by natural increase rather than immigration. Migration flows are shaped by external pressures rather than population size, highlighting the complexity of demographic dynamics in the region. This project demonstrates the value of combining exploratory data analysis with visualization tools like Power BI to uncover meaningful insights on popuation and migration.
